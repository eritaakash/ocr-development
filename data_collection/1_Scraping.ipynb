```python
from selenium import webdriver
from selenium.webdriver.common.by import By

import pandas as pd 
import time 


data = {
    'name': [],
    'url': []
}

num_pages = 39
driver = webdriver.Chrome()

for page in range(1, num_pages + 1):
    driver.get(f'https://www.dafont.com/theme.php?cat=603&page={page}&fpp=50&l[]=10&l[]=1')
    time.sleep(2)

    # find all a's having dl as classname 
    links = driver.find_elements(By.CSS_SELECTOR, 'a.dl')
    print(f'[>] Page {page} | {len(links)} links')

    for link in links:
        url = link.get_attribute('href')
        name = url.split('f=')[1]

        data['name'].append(name)
        data['url'].append(url)

        df = pd.DataFrame(data)
        df.to_csv('fonts_data.csv', index=False)
    
    time.sleep(2)
```

```python
import os
import requests
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

# Load the font data
fonts_df = pd.read_csv('./fonts.csv')

# Ensure the download directory exists
os.makedirs('./downloads', exist_ok=True)

def download_font(row):
    """Downloads a font zip file given a row from the DataFrame."""
    idx, font_name, font_url = row
    try:
        response = requests.get(font_url, timeout=10)
        if response.status_code == 200:
            with open(f'./downloads/{font_name}.zip', 'wb') as f:
                f.write(response.content)
            print(f'[{idx}] Font {font_name} downloaded')
        else:
            print(f'[{idx}] Failed to download {font_name} - HTTP {response.status_code}')
    except requests.RequestException as e:
        print(f'[{idx}] Error downloading {font_name}: {e}')


max_threads = os.cpu_count() or 8
with ThreadPoolExecutor(max_threads) as executor:
    executor.map(download_font, fonts_df.itertuples(index=True, name=None))
```

```python
import zipfile 

fonts_dir = './fonts'

if not os.path.exists(fonts_dir):
    os.makedirs(fonts_dir, exist_ok=True)

font_zips = [f for f in os.listdir('./downloads') if f.endswith('.zip')]

i = 0
for font_file in font_zips:
    font_path = f'./downloads/{font_file}'

    with zipfile.ZipFile(font_path, 'r') as zip_ref:

        extracted_files = [ f for f in zip_ref.namelist() if f.lower().endswith('.ttf') or f.lower().endswith('.otf') ]
        zip_ref.extractall(fonts_dir, members=extracted_files)

    print(f'[{i}] Extracted {len(extracted_files)} files from {font_file}')
    i += 1
```
